## <center>Config</center>
---

```python
TRAINING_1 = False # Si true, le training sera réalisé sur 1 simulation.
TRAINING_5000 = False # Si true, le training sera réalisé sur 5000 simulations.
TRAINING_180000 = False # Si true, le training sera réalisé sur 180000 simulations.
RUN_1 = False # Si false, la simulation sera réalisée sur la qmatrix de 1 simulation.
RUN_5000 = False # Si false, la simulation sera réalisée sur la qmatrix des 5000 simulations.
RUN_180000 = True # Si false, la simulation sera réalisée sur la qmatrix des 180000 simulations.
```

## <center>Imports</center>
---

```python
from QLearning import *
```

## <center>Training</center>
---

## Training with 1 simulations

```python
if TRAINING_1:
    qlearning = QLearning()

    qlearning.learn_q(1)
    qlearning.make_matrix_converge()
    qlearning.saveQmatrix('./q_matrix_1.json')
```

## Training with 5000 simulations

```python
if TRAINING_5000:
    qlearning = QLearning()

    qlearning.learn_q(5000)
    qlearning.make_matrix_converge()
    qlearning.saveQmatrix('./q_matrix_5000.json')
```

## Training with 180,000 simulations

```python
if TRAINING_180000:
    qlearning = QLearning()

    qlearning.learn_q(180000)
    qlearning.make_matrix_converge()
    qlearning.saveQmatrix('./q_matrix_180000.json')
```

## <center>Load Qmatrix</center>
---

## Load Q_matrix 1

```python
if RUN_1:
    qlearning = QLearning()
    qlearning.loadQmatrix("./q_matrix_1.json")
```

## Load Q_matrix 5000

```python
if RUN_5000:
    qlearning = QLearning()
    qlearning.loadQmatrix("./q_matrix_5000.json")
```

## Load Q_matrix 180000

```python
if RUN_180000:
    qlearning = QLearning()
    qlearning.loadQmatrix("./q_matrix_180000.json")
```

## <center>Run Model</center>

```python
qlearning.run(500)
qlearning.make_gif()
```
